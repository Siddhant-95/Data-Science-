## Books

```python
books = pd.read_csv('book.csv')
```

```python
books
```

### Apriori Algorithm

```python
books_frequent1 = apriori(books, min_support = 0.1, use_colnames = True)
books_frequent1
```

```python
books_frequent2 = apriori(books, min_support = 0.15, use_colnames = True)
books_frequent2
```

```python
books_frequent3 = apriori(books, min_support = 0.2, use_colnames = True)
books_frequent3
```

```python
books_frequent4 = apriori(books, min_support = 0.05, use_colnames = True)
books_frequent4
```

### Association Rules

```python
books_rules1 = association_rules(books_frequent1, metric = 'confidence', min_threshold = 0.5)
books_rules1
```

```python
books_rules2 = association_rules(books_frequent1, metric = 'confidence', min_threshold = 0.7)
books_rules2
```

```python
books_rules3 = association_rules(books_frequent1, metric = 'lift', min_threshold = 1)
books_rules3
```

```python
books_rules4 = association_rules(books_frequent1, metric = 'lift', min_threshold = 2)
books_rules4
```

### Visualisation

```python
support1 = books_rules1.support
confidence1 = books_rules1.confidence
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(support1, confidence1)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
support2 = books_rules2.support
confidence2 = books_rules2.confidence
```

```python
plt.scatter(support2, confidence2)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
plt.scatter(support3, confidence3)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
plt.scatter(support4, confidence4)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```
