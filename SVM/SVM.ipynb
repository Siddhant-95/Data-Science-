## Forest Fire

```python
forest = pd.read_csv('forestfires.csv')
```

```python
forest
```

```python
le = LabelEncoder()
forest.month = le.fit_transform(forest.month)
forest.day = le.fit_transform(forest.day)
```

```python
forest
```

```python
forest.info()
```

```python
X = forest.iloc[:,0:10]
Y = forest.iloc[:,30]
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 10)
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

### Classification using SVM

```python
clf = SVC(kernel = 'linear')
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)*100
print('Accuracy=', acc)
```

```python
clf = SVC(kernel = 'rbf', gamma = 1)
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)*100
print('Accuracy=', acc)
```

#### Grid Search

```python
clf = SVC()
param_grid = [{'kernel':['rbf'], 'gamma':[100,50,10,25,1,0.5,0.1,0.01,0.001,0.0001]}]
gsv = GridSearchCV(clf, param_grid, cv = 10)
gsv.fit(X_train, Y_train)
```

```python
gsv.best_params_, gsv.best_score_
```

```python
clf = SVC(C = 1, gamma = 0.0001, kernel ='rbf')
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)*100
print('Accuracy=', acc)
```

## Salary Data

### Preparing Training Data

```python
Salary_train = pd.read_csv('SalaryData_Train(1).csv')
```

```python
Salary_train
```

```python
Salary_train.Salary.value_counts()
```

```python
Salary_train.info()
```

```python
le = LabelEncoder()
Salary_train.workclass = le.fit_transform(Salary_train.workclass)
Salary_train.maritalstatus = le.fit_transform(Salary_train.maritalstatus)
Salary_train.occupation = le.fit_transform(Salary_train.occupation)
Salary_train.relationship = le.fit_transform(Salary_train.relationship)
Salary_train.race = le.fit_transform(Salary_train.race)
Salary_train.sex = le.fit_transform(Salary_train.sex)
Salary_train.native = le.fit_transform(Salary_train.native)
Salary_train.Salary = le.fit_transform(Salary_train.Salary)
```

```python
Salary_train
```

```python
Salary_train.drop('education', axis = 1, inplace = True)
```

```python
Salary_train.info()
```

### Preparing Test Data

```python
Salary_test = pd.read_csv('SalaryData_Test(1).csv')
```

```python
Salary_test
```

```python
le = LabelEncoder()
Salary_test.workclass = le.fit_transform(Salary_test.workclass)
Salary_test.maritalstatus = le.fit_transform(Salary_test.maritalstatus)
Salary_test.occupation = le.fit_transform(Salary_test.occupation)
Salary_test.relationship = le.fit_transform(Salary_test.relationship)
Salary_test.race = le.fit_transform(Salary_test.race)
Salary_test.sex = le.fit_transform(Salary_test.sex)
Salary_test.native = le.fit_transform(Salary_test.native)
Salary_test.Salary = le.fit_transform(Salary_test.Salary)
```

```python
Salary_test.drop('education', axis = 1, inplace = True)
```

```python
Salary_test
```

```python
Salary_test.info()
```

### Classification using SVM

```python
X_train = Salary_train.iloc[:,0:11]
Y_train = Salary_train.iloc[:,12]
```

```python
X_test = Salary_test.iloc[:,0:11]
Y_test = Salary_test.iloc[:,12]
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
clf = SVC(kernel = 'rbf', gamma = 0.001)
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)*100
print('Accuracy=',acc)
```

```python
confusion_matrix(Y_test, Y_pred)
```

### Grid Search

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[0.01,0.001,0.0001], 'C':[5,3,2,1,0.1]}]
gsv = GridSearchCV(clf, param_grid, cv = 3)
gsv.fit(X_train, Y_train)
```

```python
gsv.best_params_, gsv.best_score_
```
